```python
import pandas as pd
#from nltk import tokenize
from nltk.tokenize import word_tokenize
```

```python
#assuming that the twitter tweets are in a file called Tweets.csv 
#column "text" is the column of interest
twitterdf=pd.read_csv("Tweets.csv")
```

```python
twitterdf.tail(10)
```

```python
#assuming that the racial slur words are given in a list "racialslurr" 
racialslur=["bad","sad","bullying","annoying","black","not good"]
```

```python
#check if there are any null values in the column "text"
twitterdf.text.isnull().sum()
```

```python
#drop the null row as it is only one
twitterdf.dropna()
```

```python
#replace the null value, here am biasing it "good"
twitterdf["text"].fillna("good", inplace = True)
```

```python
#add a new column to file to update the d
twitterdf['profanityDegree']=" "
twitterdf
```

```python
for i in range(twitterdf.shape[0]):
    token=word_tokenize(twitterdf.text[i].lower())
    degree_of_profanity=sum(1 for t in token if t in racialslur)/ len(token)
    twitterdf['profanityDegree'][i]=degree_of_profanity
```
